## **1. Comprendre BERT et XLM-RoBERTa**

### Introduction aux modèles de transformateurs

Les modèles de Transformers sont une architecture de réseau neuronal qui a révolutionné le traitement du langage naturel (NLP). Ils sont particulièrement efficaces pour gérer les données séquentielles, comme le texte, grâce à leur mécanisme d'auto-attention. Ce mécanisme permet au modèle de peser l'importance des différentes parties de la séquence d'entrée lors du traitement de chaque élément.

### BERT (Bidirectional Encoder Representations from Transformers)

BERT est un modèle de Transformers pré-entraîné développé par Google. Il est **bidirectionnel**, c'est-à-dire qu'il prend en compte le contexte des mots à la fois à gauche et à droite pour comprendre le sens d'un mot. BERT est entraîné sur deux tâches principales :

1.  **Modélisation du langage masqué (Masked Language Model - MLM)** : Le modèle prédit des mots masqués aléatoirement dans une séquence.
2.  **Prédiction de la phrase suivante (Next Sentence Prediction - NSP)** : Le modèle prédit si deux phrases données sont consécutives ou non.

BERT a plusieurs versions pré-entraînées (par exemple, `bert-base-uncased`, `bert-large-cased`) qui varient en taille et en type de corpus d'entraînement.

### XLM-RoBERTa (Cross-lingual Language Model - Robustly Optimized BERT Pretraining Approach)

XLM-RoBERTa est une version multilingue de RoBERTa, qui est lui-même une version optimisée de BERT. L'objectif principal de XLM-RoBERTa est de construire un modèle de langage qui fonctionne bien pour de nombreuses langues simultanément. Il est entraîné sur un très grand corpus multilingue (plus de 100 langues) et n'utilise pas la tâche NSP. Au lieu de cela, il se concentre uniquement sur la tâche MLM, mais sur des corpus de texte beaucoup plus importants.

L'avantage de XLM-RoBERTa est sa capacité à transférer des connaissances entre les langues, ce qui le rend très utile pour des tâches basées sur la langue où les ressources sont limitées.

## **2. Tokenisation du texte**

La tokenisation est le processus de décomposition d'un texte en unités plus petites appelées "jetons" (tokens). Pour les modèles de Tranformers comme BERT et XLM-RoBERTa, cela implique souvent l'utilisation de sous-mots pour gérer les mots inconnus et réduire la taille du vocabulaire.

Les tokeniseurs de **transformers** ajoutent également des jetons spéciaux au début et à la fin des séquences (par exemple, **[CLS]** Et **[SEP]** pour BERT) et génèrent des masques d'attention et des IDs de type de segment nécessaires à l'entrée du modèle.

```python
# Installation de la library transformers s'il n'est pas encore installer
!pip install transformers

# Importation des library necessaires
from transformers import BertTokenizer, XLMRobertaTokenizer
```

```python
# Initialisation des tokens
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
xlm_roberta_tokenizer = XLMRobertaTokenizer.from_pretrained('xlm-roberta-base')

print("Tokenizers initialized.")
```

### Tokenisation d'une phrase unique

Nous allons tokeniser une phrase simple et observer les **input_ids**, **attention_mask**.

```python
sentence = "Yo, i'm verry happy today?"

# Tokenisation avec la fonction tokeniser de BERT
bert_encoding = bert_tokenizer(
    sentence,
    add_special_tokens=True,  # Add '[CLS]' and '[SEP]'
    max_length=64,           # Pad/truncate to a fixed length
    padding='max_length',    # Pad to max_length
    return_attention_mask=True, # Return attention mask
    return_tensors='pt'      # Return PyTorch tensors
)

print("BERT Tokenization (single sentence):")
print(f"  Input IDs: {bert_encoding['input_ids']}")
print(f"  Attention Mask: {bert_encoding['attention_mask']}")
print(f"  Decoded: {bert_tokenizer.decode(bert_encoding['input_ids'][0])}")

print("\n" + "="*50 + "\n")

# Tokenisation avec la fonction tokeniser de Roberta
xlm_roberta_encoding = xlm_roberta_tokenizer(
    sentence,
    add_special_tokens=True,  # Add '<s>' and '</s>'
    max_length=64,           # Pad/truncate to a fixed length
    padding='max_length',    # Pad to max_length
    return_attention_mask=True, # Return attention mask
    return_tensors='pt'      # Return PyTorch tensors
)

print("XLM-RoBERTa Tokenization (single sentence):")
print(f"  Input IDs: {xlm_roberta_encoding['input_ids']}")
print(f"  Attention Mask: {xlm_roberta_encoding['attention_mask']}")
print(f"  Decoded: {xlm_roberta_tokenizer.decode(xlm_roberta_encoding['input_ids'][0])}")
```

### Tokenisation de deux phrases

Nous allons maintenant tokeniser une paire de phrases. Pour BERT, cela inclura également `token_type_ids` pour distinguer les deux phrases.

```python
sentence1 = "This dog is my dog."
sentence2 = "Why he follow you?."

# Tokenize with BERT tokenizer for two sentences
bert_pair_encoding = bert_tokenizer(
    sentence1,
    sentence2,
    add_special_tokens=True,
    max_length=64,
    padding='max_length',
    return_attention_mask=True,
    return_token_type_ids=True, # Return token type IDs for BERT
    return_tensors='pt'
)

print("BERT Tokenization (two sentences):")
print(f"  Input IDs: {bert_pair_encoding['input_ids']}")
print(f"  Attention Mask: {bert_pair_encoding['attention_mask']}")
print(f"  Token Type IDs: {bert_pair_encoding['token_type_ids']}") # 0 for first sentence, 1 for second
print(f"  Decoded: {bert_tokenizer.decode(bert_pair_encoding['input_ids'][0])}")

print("\n" + "="*50 + "\n")

# Tokenize with XLM-RoBERTa tokenizer for two sentences
# XLM-RoBERTa typically doesn't use token_type_ids for sentence separation like BERT
xlm_roberta_pair_encoding = xlm_roberta_tokenizer(
    sentence1,
    sentence2,
    add_special_tokens=True,
    max_length=64,
    padding='max_length',
    return_attention_mask=True,
    return_tensors='pt'
)

print("XLM-RoBERTa Tokenization (two sentences):")
print(f"  Input IDs: {xlm_roberta_pair_encoding['input_ids']}")
print(f"  Attention Mask: {xlm_roberta_pair_encoding['attention_mask']}")
print(f"  Decoded: {xlm_roberta_tokenizer.decode(xlm_roberta_pair_encoding['input_ids'][0])}")
```

## **3. Préparation des données d'entrée pour le modèle**

Pour que les modèles de transformateurs puissent traiter le texte correctement, les entrées doivent être dans un format spécifique, ce qui implique plusieurs étapes clés :

### Token Spéciaux

Les modèles Transformers utilisent des tokens spéciaux pour signaler le début et la fin d'une séquence, ainsi que pour séparer plusieurs segments de texte (comme deux phrases). Ces tokens sont essentiels pour le bon fonctionnement des modèles.

*   **Pour BERT :**
    *   **[CLS]** (Class Token) : C'est le premier token de chaque séquence. L'état caché correspondant à ce token est souvent utilisé comme représentation agrégée de l'ensemble de la séquence pour des tâches de classification.
    *   **[SEP]** (Separator Token) : Utilisé pour indiquer la fin d'une phrase ou pour séparer deux phrases différentes dans une entrée.

*   **Pour XLM-RoBERTa :**
    *   **`<s>`** (Sentence Start) : Marque le début d'une séquence ou d'une phrase.
    *   **`</s>`** (Sentence End) : Marque la fin d'une séquence ou d'une phrase. Lorsqu'il y a deux phrases, un **</s>** est inséré entre elles et un autre à la fin de la seconde phrase.

### Masque d'Attention (`attention_mask`)

Le **attention_mask** est une séquence de 0 et de 1 qui indique au modèle quels jetons doivent être pris en compte (1) et quels jetons doivent être ignorés (0). Ceci est crucial pour le remplissage (`padding`) : lorsque des séquences sont rendues de la même longueur en ajoutant des tokens de remplissage, l'**attention_mask** garantit que le modèle ne porte pas attention à ces tokens de remplissage, qui n'ont aucune signification sémantique.

### Longueur Maximale et Remplissage/Troncation

Les modèles de Transformers ont une longueur d'entrée maximale fixe (par exemple, 512 tokens pour la plupart des versions de BERT). Les phrases plus courtes sont complétées avec des tokens de remplissage (**[PAD]** ou **<pad>**) jusqu'à cette longueur maximale, et les phrases plus longues sont tronquées. L'attention_mask gère les tokens de remplissage.

```python
# Afficher les tokens spéciaux pour BERT
print("BERT Special Tokens Map:")
print(bert_tokenizer.special_tokens_map)
print(f"BERT Vocabulary Size: {bert_tokenizer.vocab_size}")

print("\n" + "="*50 + "\n")

# Afficher les tokens spéciaux pour XLM-RoBERTa
print("XLM-RoBERTa Special Tokens Map:")
print(xlm_roberta_tokenizer.special_tokens_map)
print(f"XLM-RoBERTa Vocabulary Size: {xlm_roberta_tokenizer.vocab_size}")
```

## 5. Création de plis de validation croisée

Pour évaluer la performance de notre modèle de manière robuste et éviter le surapprentissage, nous allons utiliser une validation croisée à k plis. **StratifiedKFold** est choisi pour maintenir la proportion des classes dans chaque pli, ce qui est crucial pour les tâches de classification.

Nous allons créer 5 plis (**n_splits=5**), mélanger les données (**shuffle=True**) pour garantir la randomisation, et fixer un **random_state** pour la reproductibilité.

```python
from sklearn.model_selection import StratifiedKFold

# Initialisation de StratifiedKFold
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Listes pour stocker les DataFrames d'entraînement et de validation pour chaque pli
train_folds = []
val_folds = []

print("Génération des plis de validation croisée...")

# Split the training data into folds
for fold, (train_index, val_index) in enumerate(kf.split(df_train, df_train['label'])):
    print(f"Processing Fold {fold+1}/")

    # Create train and validation sets for the current fold
    fold_train_df = df_train.iloc[train_index].reset_index(drop=True)
    fold_val_df = df_train.iloc[val_index].reset_index(drop=True)

    train_folds.append(fold_train_df)
    val_folds.append(fold_val_df)
```

```python

print("Cross-validation folds created successfully!")
print(f"Number of training folds: {len(train_folds)}")
print(f"Number of validation folds: {len(val_folds)}")
```

```python

# Display shape of the first training and validation fold
print(f"Shape of first training fold: {train_folds[0].shape}")
print(f"Shape of first validation fold: {val_folds[0].shape}")
```

```python

# Display the label distribution in the first fold to verify stratification
print("\nLabel distribution in first training fold:")
display(train_folds[0]['label'].value_counts(normalize=True))
```

```python
print("\nLabel distribution in first validation fold:")
display(val_folds[0]['label'].value_counts(normalize=True))
```
