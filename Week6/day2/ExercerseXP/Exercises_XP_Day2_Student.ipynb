### BERT
- Transformer bidirectionnel à encodeur uniquement ; entraîné avec la modélisation du langage masqué (MLM) et historiquement la prédiction de la phrase suivante (NSP)
- Application: Reconnaissance d'entités nommées (NER), analyse des sentiments, réponse aux questions
- Why: BERT comprend le contexte venant de la gauche et de la droite, ce qui le rend très efficace pour les tâches de compréhension du langage.

### GPT
- Transformer uniquement décodeur ; entraîné à l’aide d’un modèle de langage causal autorégressif;
- Application: Chatbots, génération de texte, génération de code, création de contenu
- Why: GPT excelle dans la génération de texte cohérent car il apprend à prédire et à produire des séquences un jeton à la fois.

### T5
- Transformateur encodeur-décodeur ; traite chaque tâche de NLP comme un problème de conversion texte à texte
- Application: Résumé de texte, traduction automatique, réponse aux questions
- Why: L'architecture encodeur-décodeur est parfaitement adaptée à la transformation d'une séquence de texte en une autre, ce qui la rend flexible pour de nombreuses tâches de traitement automatique du langage naturel.


## 🌟 Exercise 3: Benefits and Ethics of Pre-training
- Benefits (explain each in your words):
  1) Improved generalization: Les modèles pré-entraînés apprennent les schémas linguistiques généraux à partir de vastes ensembles de données, ce qui leur permet d'obtenir de bons résultats sur des tâches nouvelles et inédites avec une meilleure précision.
  2) Less labeled data: Étant donné que le modèle acquiert déjà des connaissances linguistiques lors du pré-entraînement, il nécessite moins d'exemples étiquetés manuellement pour obtenir de bonnes performances sur les tâches en aval.
  3) Faster fine-tuning: Un modèle pré-entraîné comprend déjà les structures du langage, son adaptation à une tâche spécifique nécessite donc moins de temps d'entraînement et de ressources informatiques.
  4) Transfer learning: Les connaissances acquises lors de la formation préalable peuvent être transférées à différentes tâches, telles que la classification, la traduction, la synthèse et la réponse aux questions.
  5) Robustness: L'exposition à des données diverses pendant la phase de pré-entraînement aide le modèle à gérer plus efficacement les entrées bruitées, les fautes d'orthographe, le langage informel et les variations de domaine.

- Ethical concerns: Les modèles pré-entraînés peuvent hériter de biais issus des données d'entraînement, générer des informations erronées, être utilisés à des fins malveillantes ou exposer des informations sensibles.
- Mitigations:
  * Gestion des données : Supprimer les données biaisées et de faible qualité.
  * Confidentialité différentielle : Protéger les informations sensibles pendant la formation.
  * Filtres de sécurité : bloquent les sorties nocives ou dangereuses.
  * Audits : Évaluer régulièrement les modèles afin de déceler les biais, l’équité et les problèmes de sécurité.


## 🌟 Exercise 4: Transformer Architecture Deep Dive
### Self-Attention & Multi-Head Attention
- TODO: Describe the Q/K/V flow, softmax weighting, and value mixing.
  * Chaque jeton est transformé en trois vecteurs : requête (Q) , clé (K) et valeur (V) .
  * La requête d'un jeton est comparée aux clés de tous les jetons à l'aide de produits scalaires pour calculer les scores d'attention.
  * Ces scores sont mis à l'échelle et traités par une fonction softmax pour obtenir les pondérations d'attention.
  * Les pondérations déterminent l'attention à accorder à chaque jeton.
  * La représentation finale est obtenue en calculant une somme pondérée des vecteurs de valeur.
- TODO: Explain why multiple heads help (subspace projections, diverse relations).
  * Chaque tête d'attention apprend différentes projections de l'entrée dans des sous-espaces distincts.
  * Différentes têtes peuvent se concentrer simultanément sur différents types de relations.
  * Cela permet au modèle de saisir les dépendances syntaxiques, sémantiques et contextuelles plus efficacement qu'un mécanisme d'attention unique.

### Pre-training Objectives
- MLM vs CLM: TODO (compare objectives and typical use)
  * **MLM (Modélisation du langage masqué)** : Des mots aléatoires sont masqués et le modèle prédit les mots manquants, utilise le contexte gauche et droit (exemple : Bert)
  * **Modélisation causale du langage (CLM)** : Prédit le prochain jeton en se basant uniquement sur les jetons précédents et utilise le contexte de gauche à droite (GPT)
- When to prefer MLM vs CLM:
 * **MLM** : Plus adapté aux tâches de compréhension du langage telles que la classification, la reconnaissance d’entités nommées et l’analyse des sentiments.
 * **CLM** : Meilleur pour la génération de texte, les systèmes de dialogue, la génération de code et la création de contenu.
- NSP: TODO (why early BERT used it and why many modern models avoid it)
  * Utilisé dans les premières versions de BERT pour déterminer si une phrase suit logiquement une autre.
  * Destiné à améliorer la compréhension au niveau de la phrase et les performances en matière de réponse aux questions.

### Transformer Model Selection
- Sentiment on reviews:
  * Encodeur : BERT
  * Justifcation : L’analyse des sentiments nécessite de comprendre l’ensemble du contexte de l’avis plutôt que de générer du texte.
- Conversational chatbot
  * Décodeur : GPT
  * Justification : Les chatbots doivent générer des réponses cohérentes et fluides, token par token.
- Technical document
  * Encodeur-Décodeur : T5
  * Justification : La traduction nécessite l’encodage de la phrase source et la génération d’une phrase cible correspondante.

### Positional Encoding
- Purpose:
  * Les transformers traitent tous les jetons en parallèle et ne connaissent pas intrinsèquement l'ordre des mots.
  * Les encodages positionnels fournissent des informations sur la position de chaque jeton dans une séquence.
  * Ils aident le modèle à distinguer les mots apparaissant à différents endroits et à comprendre la structure séquentielle.
- Example failure without positions:
«Le chat a poursuivi le chien.»
«Le chien a poursuivi le chat.»  
Sans information de position, les deux phrases contiennent les mêmes mots et pourraient paraître identiques au modèle, rendant difficile de déterminer qui poursuit qui. L'encodage positionnel préserve l'ordre et le sens de la phrase.

## 🌟 Exercise 6: Softmax Temperature: The Randomness Regulator
### 1) Temperature Scenarios
- T=0.2: Le modèle privilégie nettement les tokens les plus probables et les résultats sont prévisibles
- T=1.5: Résultat aléatoire et le modèle renvoie des résultats plus diversifiés et créatifs, mais peut générer des textes non pertinents ou incohérents.
- T=1.0: équilibre entre créativité et cohérence et les résultats sont généralement naturels et diversifiés sans devenir erratiques

### 2) Application Design
- Bedtime stories (creativity vs coherence): TODO (temperature strategy and why)
Température recommandée : 0,8 – 1,2  
Pourquoi:
  * Les histoires du soir doivent être imaginatives et captivantes.
  * Une température modérée à élevée favorise la créativité et un vocabulaire varié.
  * Maintenir la température en dessous des valeurs extrêmes contribue à préserver la cohérence de l'intrigue et des personnages.
- Financial report summaries (accuracy/reliability): TODO (temperature/decoding strategy and why)
Température recommandée : 0,0 – 0,3  
Stratégie de décodage : décodage glouton ou échantillonnage à basse température.  
Pourquoi:
  * Les rapports financiers exigent une cohérence et une précision factuelles.
  * Les basses températures réduisent le caractère aléatoire et les hallucinations.
  * Le modèle est plus susceptible de sélectionner la formulation la plus probable et la plus fiable.

### 3) Temperature & Bias
- TODO: How does temperature influence bias surfacing or dampening? Give a realistic example.

Comment la température influence les biais :

La température ne crée pas de biais en soi. Les biais proviennent des données d'entraînement et des paramètres du modèle. Cependant, la température peut influencer la fréquence d'apparition des résultats biaisés.

* **Basse température** : le modèle tend à privilégier les réponses les plus probables. Si une association biaisée a une probabilité élevée, elle risque d’apparaître plus fréquemment.
* **Température élevée** : Le modèle explore un éventail de possibilités plus large. Cela peut parfois réduire les réponses biaisées répétitives, mais aussi faire émerger des stéréotypes moins courants ou des associations problématiques.

Exemple : Une infirmière est entrée dans la pièce. Qu'a-t-elle fait ensuite ?

À basse température (0,2) , le modèle peut générer de manière répétée un schéma commun appris à partir des données, tel que :

-> Elle a vérifié l'état du patient.

À une température plus élevée (1,5) , le modèle peut produire des continuations plus diversifiées :

-> L'infirmière a examiné le dossier.

-> L’infirmière a expliqué le plan de traitement.

-> Il a préparé les médicaments.
