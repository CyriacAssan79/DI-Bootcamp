## **Exercice 5**

```python
import matplotlib.pyplot as plt

np.random.seed(0)
n_days = 15
days = np.arange(n_days)
ts_values = np.random.rand(n_days)  # valeurs aléatoires

# Ajustement d'une droite de tendance linéaire
slope, intercept = np.polyfit(days, ts_values, 1)
trend = slope * days + intercept

# Coefficient de corrélation de Pearson
pearson_r = np.corrcoef(days, ts_values)[0, 1]

# Affichage / tracé
plt.figure(figsize=(8, 4))
plt.plot(days, ts_values, 'o-', label='Série temporelle')
plt.plot(days, trend, 'r--', label=f'Droite de tendance (pente={slope:.3f})')
plt.xlabel('Jour')
plt.ylabel('Valeur')
plt.title('Série temporelle sur 15 jours et droite de tendance')
plt.legend()
plt.grid(True)
plt.show()

print(f"Coefficient de corrélation de Pearson (jours vs valeurs) : {pearson_r:.4f}")

# Formule du coefficient de corrélation de Pearson :
# r = sum((x - mean(x)) * (y - mean(y))) / ( sqrt(sum((x - mean(x))**2)) * sqrt(sum((y - mean(y))**2)) )
```
