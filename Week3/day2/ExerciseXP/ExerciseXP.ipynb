## **Exercice 5**

Une image en niveaux de gris est représentée par un tableau 2D ou chaque élément du tableau correspond à un pixel. De plus, la valeur du pixel représente son intensité lumineuse et généralement 0 pour noir et 255 pour blanc.  
Les valeurs intermédiaires représentent différents niveaux de gris.

```python
#Exemple
image = np.array([
    [0, 50, 100, 150, 200],
    [25, 75, 125, 175, 225],
    [50, 100, 150, 200, 250],
    [75, 125, 175, 225, 255],
    [100, 150, 200, 225, 255]
])

print(image)
```

```python
#Affichage de l'image

plt.imshow(image, cmap='gray')
plt.title("Image 5x5 en niveaux de gris")
plt.colorbar()
plt.show()
```

Quand au image de couleur, ils sont représenter par un tableau 3D et généralement avec trois canaux (Rouge, Vert, Bleu).  
Par exemple, une image couleur de 100 × 100 pixels aura une forme :(100, 100, 3)

Chaque pixel contient alors trois valeurs :

[255, 0, 0]     # Rouge  
[0, 255, 0]     # Vert  
[0, 0, 255]     # Bleu  
[255, 255, 255] # Blanc  
[0, 0, 0]       # Noir  

Dans NumPy, une image est simplement un tableau numérique :  

* 2D pour une image en niveaux de gris.  
* 3D pour une image couleur (RGB).  

## **Exercice 6**

```python
import numpy as np

# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)
```

```python
productivity_before
```

```python
productivity_after
```

```python
productivity_after.mean()
```

```python
productivity_before.mean()
```

H0 : Les deux moyenne sont égales  
H1 : Les deux moyenne sont différentes

```python
# Performing a one-sample t-test
t_statistic, p_value = ttest_rel(productivity_before, productivity_after)

# Interpreting the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
p_value
```

Vu que p-value < 0.05, on rejette l'hypothèse null H0, donc il existe une différence statistiquement significative entre la productivité moyenne avant et après la formation.  
Les résultats suggèrent que le programme de formation a eu un effet significatif sur la productivité des employés.
