## **Exercice 2**

```python
t2 = np.random.rand(20)
t2 = np.round(t2, 2)
```

```python
t3 = np.random.rand(20)
t3 = np.round(t3, 2)
```

```python
commun = []
```

```python
for i in t2:
    for j in t3:
        if i == j:
            commun.append(i)
```

```python
commun
```
