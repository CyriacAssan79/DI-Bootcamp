```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## **1. Data Preparation:**

```python
np.random.seed(42)
temp = np.random.uniform(low=-5, high=35, size=(10,12))
```

Nous génerons des nombres aléatoires uniformes compris entre -5 et 35 dans une matrice 10*12(10 lignes, 10 colonnes)

```python
temp
```

On affiche nos notre matrice généré

```python
month = ['Jan','Feb','Mar', 'Apr','May', 'Jun', 'Jul'
         , 'Aug', 'Sep','Oct','Nov','Dec']
city = ['Abidjan','Yamoussoukro', 'Bouake', 'Man', 'Korhogo','Divo','Toumodi',
        'Gagnoa','Sassandra','Agboville']
```

```python
df = pd.DataFrame(data=np.round(temp,2), index=city, columns=month)
```

On stocke notre matrice généré arrondi à l'odre 2, la ville comme index et les mois comme colonnes dans un nouveau datasets

```python
df
```

On affiche montre dataset

## **3. Visualisation des données :**

```python
plt.figure(figsize=(10,8),dpi=100)

for ville in df.index:
    plt.plot(df.columns, df.loc[ville], label=ville)

plt.title("Températures moyennes mensuelles par ville")
plt.xlabel("Mois")
plt.ylabel("Température (°C)")
plt.legend(loc= "lower left")
plt.grid(True)
plt.show()
```

Dans cette partie, on voit l'évolution de la température de chaque au cours de l'année (Janvier, Dacembre)

## **Conclusion**

Notre jeu se compose de 10 villes et les températures sont observées de janvier à décembre, c'est lors d'une année. Certaines villes comme Gagnoa ont des températures moyennes élevées contrairement à d'autres et certaines des températures moyennes relativement basses (Yamoussoukro). Cependant, pour chaque ville, les températures ont tendance à varier et ne sont pas constantes (comme le montre le graphique).
