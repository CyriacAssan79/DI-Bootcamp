## **3. Chargement du modèle**

```python
from llama_cpp import Llama

llm = Llama(
    model_path="llama2-7b-chat.gguf",
    n_gpu_layers=-1,      # utilise toutes les couches sur le GPU
    n_threads=4,
    verbose=False
)
```

## **4. Génération de texte**

```python
prompt = "Explain how the solar system formed."

output = llm(
    prompt=prompt,
    max_tokens=256
)

print(output["choices"][0]["text"])
```

## **5. Génération de code Python en streaming**

```python
output_stream = llm(
    prompt="Write a Python script that loads a Hugging Face model and tokenizes input.",
    max_tokens=256,
    stream=True
)

# Iterate through the tokens and print them at the same time they are computed

for output in output_stream:
    token = output["choices"][0]["text"]
    print(token, end="", flush=True)
```
