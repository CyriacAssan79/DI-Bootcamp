## **Loading library**

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical
```

## **Load the MNIST dataset:**

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

Il y a 60000 images dans les donnГ©es de test et 10000 images dans les donnГ©es d'entraГ®nements avec une rГ©solution de 28x28

## **Preprocess the data for a Fully Connected Neural Network:**

```python
X_train_flat = X_train.reshape((60000, 28 * 28))
X_test_flat = X_test.reshape((10000, 28 * 28))
```

```python
X_train_flat = X_train_flat.astype('float32') / 255.0 #Normalisation des images pour conserver l'essentielle
X_test_flat = X_test_flat.astype('float32') / 255.0
```

```python
y_train_ho = to_categorical(y_train) #CatГ©gorisation de la variable cible
y_test_ho = to_categorical(y_test)
```

## **Build and train a Fully Connected Neural Network:**

```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(28 * 28,)),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train_flat, y_train_ho, epochs=10, batch_size=32, validation_split=0.2) #Entrainement du modГЁle avec 10 Г©poques
```

```python
loss, accuracy = model.evaluate(X_test_flat, y_test_ho)
print(f'Test accuracy: {round(accuracy * 100,2)}%')
print(f'Test loss: {round(loss,4)}')
```

## **Preprocess the data for a Convolutional Neural Network:**

```python
X_train_cnn = X_train.reshape((60000, 28, 28, 1))
X_test_cnn = X_test.reshape((10000, 28, 28, 1))
```

```python
X_train_cnn = X_train_cnn.astype('float32') / 255.0
X_test_cnn = X_test_cnn.astype('float32') / 255.0
```

```python
print(X_train_cnn.shape)
print(X_test_cnn.shape)
```

## **Build and train a Convolutional Neural Network:**

```python
model_cnn = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),

    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model_cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model_cnn.fit(X_train_cnn, y_train_ho, epochs=10, batch_size=32, validation_split=0.2)
```

```python
loss, accuracy = model_cnn.evaluate(X_test_cnn, y_test_ho)
print(f'Test accuracy: {round(accuracy * 100,2)}%')
print(f'Test loss: {round(loss,2)}')
```

## **Compare the performance:**

Dans les deux, les modГЁles ci ont un accuracy suppГ©rieure aux ann (97% pour les ANN et 98% pour les CNN)
