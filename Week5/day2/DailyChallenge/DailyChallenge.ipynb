## **1. Understand Classification Types**

-> La classification binaire est une méthode d'apprentissage automatique supervisé qui consiste à attribuer chaque donnée à l'une des deux catégories distinctes possibles (0,1 ou Vrai, Faux etc), par exemple on veut déterminer s'il y a une fraude à la banque (1 s'il y a fraude et 0 s'il n'y pas de fraude)  
-> La classification multiclasse est une tâche d'apprentissage automatique qui consiste à attribuer une observation (comme une image, un texte ou un chiffre) à une seule catégorie parmi plus de deux classes possibles (par exemple, identifier si un animal est un chat, un chien ou un oiseau).  
-> La classification multi-label (ou multi-étiquettes) est une tâche d'apprentissage automatique supervisé où chaque donnée peut être associée simultanément à plusieurs catégories ou labels (par exemple, un film peu être à la fois un film d'action et de comédie

## **2. Set Up Your Python Environment and dataset**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import make_circles

samples = 1000
X, y = make_circles(samples,noise = 0.03,random_state = 42)

print('X : ', X)
print('\n')
print('y : ', y)
```

```python
plt.figure(figsize=(10,8),dpi=80)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

## **3. Build a Basic Neural Network Model**

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1,input_shape=(2,),activation='sigmoid') # Changed output units from 2 to 1
])

model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
    loss=tf.keras.losses.BinaryFocalCrossentropy(),
                  metrics=['accuracy'])

model.fit(X,y,epochs=10)
```

L'accuracy est encore très bas

```python
history = model.fit(
    X,
    y,
    epochs=15,
    verbose=1
)
```

```python
#Evaluation
loss, accuracy = model.evaluate(X, y)

print("Loss :", loss)
print("Accuracy :", accuracy)
```

Avec une seule couche dense, le modèle obtient généralement une précision relativement faible car les données ne sont pas linéairement séparables.

## **4. Improve the Model**

```python
model_improve = tf.keras.Sequential([
    tf.keras.layers.Dense(20,input_shape=(2,),activation='relu'),
    tf.keras.layers.Dense(10,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])

model_improve.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.BinaryFocalCrossentropy(),
                  metrics=['accuracy'])

model_improve.fit(X,y,epochs=30)
```

```python
#Evaluation
loss, accuracy = model_improve.evaluate(X, y)

print("Loss :", loss)
print("Accuracy :", accuracy)
```

Le modèle ,arrive a performer avec l'optimiseur Adam et l'ajout de couche

## **5. Visualize the Decision Boundary**

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_decision_boundary(model, X, y):

    # Définition des limites du graphique
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1

    # Création d'une grille de points
    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 200),
        np.linspace(y_min, y_max, 200)
    )

    # Prédictions sur chaque point de la grille
    x_in = np.c_[xx.ravel(), yy.ravel()]

    y_pred = model.predict(x_in, verbose=0)

    # Conversion probabilités -> classes
    y_pred = np.round(y_pred)

    # Remise en forme
    y_pred = y_pred.reshape(xx.shape)

    # Affichage de la frontière
    plt.figure(figsize=(7, 6))

    plt.contourf(xx, yy, y_pred, alpha=0.6, cmap="viridis")

    plt.scatter(X[:, 0], X[:, 1],c=y,cmap="viridis", edgecolors="black")

    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title("Decision Boundary")

    plt.show()
```

```python
plot_decision_boundary(model_improve,X,y)
```

La visualisation de la frontière de décision permet de comprendre ce que le réseau neuronal a réellement appris. Le modèle simple ne peut construire qu'une séparation presque linéaire et échoue à représenter la structure concentrique du dataset make_circles. En ajoutant des couches cachées et davantage de neurones, le réseau apprend des relations non linéaires plus complexes. La frontière de décision devient alors circulaire et s'adapte beaucoup mieux aux données, ce qui améliore significativement la précision du modèle.

## **6. Incorporate Activation Functions**

### **Explore ReLU and Sigmoid activation functions.**

#### **Sigmoid**

La fonction sigmoïde transforme une valeur en une probabilité comprise entre 0 et 1.

$\sigma(x) = \frac{1}{1 + e^{-x}}$
	​

**Caractéristiques**
* Sortie comprise entre 0 et 1.
* Idéale pour la classification binaire.
* Souvent utilisée dans la couche de sortie.

**ReLu**  
Elle est définie par :    
$ReLU(x) = \max(0, x)$  
**Caractéristiques:**  
* Très simple à calculer.
* Accélère l'entraînement.
* Réduit le problème du gradient qui disparaît.
* Très utilisée dans les couches cachées.

### **Modify the neural network to include these activation functions**



```python
model_relu_sigmoid = tf.keras.Sequential([
    tf.keras.layers.Dense(16,input_shape=(2,),activation='relu'),
    tf.keras.layers.Dense(16,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])

model_relu_sigmoid.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.BinaryFocalCrossentropy(),
                  metrics=['accuracy'])

model_relu_sigmoid.fit(X,y,epochs=30)
```

```python
result_relu_sigmoid = model_relu_sigmoid.evaluate(X, y)
```

```python
loss, accuracy = model_relu_sigmoid.evaluate(X, y)

print("Loss :", loss)
print("Accuracy :", accuracy)
```

Lorsqu'on ajoute les fonctions d'activation relu et sigmoid dans les couches, le modèle apprend performe à parmir de la 30ème époque
