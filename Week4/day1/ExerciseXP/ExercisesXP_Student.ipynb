```python

# This piece of code is already prefilled, run it to execute it and see the results.
# It demonstrates standard classification metrics for binary loan default prediction.

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, average_precision_score, confusion_matrix, classification_report

# Please replace these placeholders with your true labels and predicted probabilities.
y_true = [0,1,0,1,0,0,1,0,1,0]            # placeholder labels
y_pred_proba = [0.05,0.80,0.10,0.65,0.20,0.15,0.70,0.30,0.85,0.25]  # placeholder probabilities

# You should set a decision threshold that reflects the precision–recall trade-off for your business case.
threshold = 0.5
y_pred = [1 if p >= threshold else 0 for p in y_pred_proba]

print("Accuracy:", round(accuracy_score(y_true, y_pred), 4))
print("Precision:", round(precision_score(y_true, y_pred, zero_division=0), 4))
print("Recall:", round(recall_score(y_true, y_pred, zero_division=0), 4))
print("F1-score:", round(f1_score(y_true, y_pred, zero_division=0), 4))
print("ROC-AUC:", round(roc_auc_score(y_true, y_pred_proba), 4))
print("PR-AUC (Average Precision):", round(average_precision_score(y_true, y_pred_proba), 4))
print("\nConfusion matrix:\n", confusion_matrix(y_true, y_pred))
print("\nClassification report:\n", classification_report(y_true, y_pred, zero_division=0))
```

### Evaluation du model d'apprentissage suppervisé (e.g., Classification for Loan Default)
Pour un modèle de classification supervisée tel que la prédiction de défaut de paiement d'un prêt, une stratégie d'évaluation efficace commence par la division des données en ensembles d'entraînement, de validation et de test afin de garantir une évaluation objective des performances. Une approche courante consiste à utiliser la validation croisée k-fold sur l'ensemble d'entraînement pour optimiser les hyperparamètres et obtenir une estimation robuste des performances du modèle avant l'évaluation finale sur l'ensemble de test non utilisé lors de l'entraînement. Les métriques pertinentes à rapporter incluent l'exactitude, la précision, le rappel, le score F1, l'aire sous la courbe ROC (AUC-ROC) et l'aire sous la courbe précision-rappel (AUC-PR). Le seuil de décision, qui convertit les probabilités prédites en étiquettes de classe, est choisi en analysant la courbe précision-rappel ou la courbe ROC, en privilégiant la minimisation des faux positifs (pour la précision) ou des faux négatifs (pour le rappel), selon le coût opérationnel associé à chaque type d'erreur. L'un des principaux défis de l'évaluation des modèles supervisés consiste à s'assurer que les indicateurs choisis reflètent véritablement l'objectif commercial et que le modèle se généralise bien aux données réelles non observées, en particulier lorsque la distribution des données évolue au fil du temps.

### Evaluation du model d'apprentissage non suppervisé (e.g., Clustering)
Lors de l'évaluation d'un modèle de clustering non supervisé, tel que K-Means ou DBSCAN, l'objectif est d'évaluer la qualité et la structure des clusters formés sans étiquettes de vérité terrain. Pour mesurer la qualité des clusters, on utilise généralement des métriques intrinsèques comme le **Score de Silhouette**, qui mesure la similarité d'un objet avec son propre cluster par rapport aux autres clusters. Une autre métrique utile est l'**Indice de Davies-Bouldin**, qui quantifie le taux de similarité moyen entre chaque cluster et le cluster le plus similaire. Des graphiques de diagnostic, tels que des nuages ​​de points avec des clusters colorés, permettent de visualiser si les clusters sont bien séparés et cohérents. La **Méthode du Coude** (qui consiste à représenter graphiquement la somme des carrés intra-cluster en fonction du nombre de clusters) est utilisée pour déterminer un nombre optimal de clusters en recherchant le point d'inflexion où le taux de diminution de la distorsion ralentit significativement. L'un des principaux défis d'évaluation des modèles non supervisés réside dans l'absence de vérité terrain claire et objective, ce qui rend difficile d'affirmer avec certitude qu'un résultat de clustering est « meilleur » qu'un autre sans expertise du domaine et validation externe.

### Evaluation du model d'apprentissage par renforcement
L'évaluation d'un modèle d'apprentissage par renforcement (RL) consiste à évaluer sa capacité à apprendre une politique optimale et à être performant dans son environnement. Le succès est principalement mesuré par la **récompense cumulée** que l'agent accumule au fil du temps ; une récompense cumulée plus élevée indique une meilleure performance. Nous suivons cette récompense au fil des épisodes pour observer la progression et la convergence de l'apprentissage, en recherchant une stabilisation de la courbe de récompense, suggérant que l'agent a appris une politique stable. L'équilibre entre l'**exploration** (essayer de nouvelles actions pour découvrir de meilleures stratégies) et l'**exploitation** (utiliser la meilleure stratégie connue) est crucial et peut être mesuré en surveillant la fréquence des actions nouvelles par rapport aux actions optimales connues. Des techniques comme les algorithmes epsilon-greedy ou de borne supérieure de confiance (UCB) gèrent explicitement cet équilibre. Un défi majeur de l'évaluation des modèles d'apprentissage par renforcement réside dans la stochasticité inhérente aux environnements et la difficulté à concevoir des fonctions de récompense qui reflètent fidèlement le comportement souhaité, ce qui peut conduire les agents à trouver des solutions non intentionnelles ou à présenter un apprentissage instable.
