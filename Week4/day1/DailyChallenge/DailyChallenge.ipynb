## **Machine learning et analyse de données**

Bien que ces deux concepts aient des définitions particulières, ils sont intimement liés.  
L'analyse de données se définit comme l'ensemble des processus permettant de collecter, traiter, analyser et interpréter des informations. Toutefois, ces informations traitées peuvent servir dans plusieurs autres domaines, comme le marketing, où elles permettent de générer des rapports montrant l'évolution des ventes, ou encore pour les développeurs, afin de consulter la liste des utilisateurs. Cependant, nous pouvons pousser nos recherches afin d'utiliser ces données traitées pour prédire des événements, d'où la nécessité du machine learning. En effet, chaque modèle s'établit sur des caractéristiques pour pouvoir prédire un résultat.

## **Secteur concerné par le machine learning**

Le machine learning intervient dans plusieurs secteurs : 

* **Prédiction du prix d'une maison** : Dans l'immobilier, on peut utiliser le machine learning pour prédire le prix d'une maison en fonction de plusieurs caractéristiques comme les dimensions de la maison, le nombre de pièces, le quartier, etc. Le modèle se base sur ces features pour prédire le prix de la maison.

* **Prédiction de la catégorie d'une fleur** : la prédiction de la catégorie d'une fleur d'iris est l'un des exemples les plus utilisés dans le domaine du machine learning. En effet, le modèle reçoit 4 variables d'entrée (longueur pétale, largeur pétale, longueur sépale, largeur sépale), puis prédit la catégorie à laquelle appartient cette fleur.

* **Prédiction de zone à risque** : on peut utiliser le machine learning dans la sécurité routière, avec la prédiction des zones à risque d'accidents grâce à certaines caractéristiques comme la concentration de population dans les environs, les conditions météorologiques, le type de la route, etc. Puis le modèle se base sur ça pour prédire le résultat.

## **Apprentissage supervisé # apprentissage non suppervisé # apprentissage par renforcement**

Pour celà, nous nous baserons sur des exemples :  

* **Apprentissage supervisé** : supposons qu'on veut évaluer un élève dans la matière de maths, l'élève a ses réponses qu'il a données et derrière il y a des réponses exactes. C'est un peu comme ça que l'apprentissage supervisé fonctionne : on lui donne des features et des résultats (labels) puis il s'entraine sur ça pour pouvoir prédire le résultat.  

* **Apprentissage non supervisé** : supposons qu'on a plusieurs bonbons, on a des bonbons de couleurs différentes, de formes différentes, etc. Le modèle n'a pas de label, donc il groupe les éléments en fonction de leurs similarités.

* **Apprentissage par renforcement** : supposons qu'on a un robot qui doit apprendre à marcher comme un nouveau-né. À chaque fois qu'il essaie, il tombe, puis répète les mêmes choses et réussit finalement à marcher. C'est un peu comme ça que l'apprentissage par renforcement fonctionne : le modèle essaie plusieurs fois une liste d'actions afin d'atteindre le but qu'on lui a fixé.

## **Processus de développement d'un modèle d'apprentissage automatique**

Lorsqu'on veut développer un algorithme de machine learning, l'une des premières étapes est la sélection des features. En effet, il peut arriver que lors de la collecte on se retrouve avec une trentaine ou une cinquantaine de features et dans ces features nous pouvons avoir des informations qui ne sont pas pertinentes pour la prédiction, je peux prendre le cas du nom d'un utilisateur ou de son ID, donc il est essentiel de parcourir ces features afin de sélectionner celles idéales pour notre modèle de prédiction. Ensuite vient l'étape de sélection de l'algorithme idéal pour le problème, et pour cela, nous avons plusieurs types d'algorithmes de machine learning (supervisé, non supervisé, renforcement, etc.). En fonction de ce qu'on veut, chaque modèle possède des algorithmes de machine learning qui lui sont dédiés, par exemple supervisé avec la régression linéaire ou logistique, non supervisé, plus dans le cas du deep learning, et le renforcement comme le Q-learning. Enfin, après avoir utilisé les algorithmes adéquats, il faut les tester avec des métriques comme l'accuracy, la precision, le recall, etc. Chaque modèle a des métriques de validation adéquates.
