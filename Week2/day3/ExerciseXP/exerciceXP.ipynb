## **Exercice 3**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

norm = norm(loc = 50, scale = 10)
print("PDF at x=0.5:", norm.pdf(50))
```

```python
import matplotlib.pyplot as plt

# Plotting Gaussian Distribution
x_gaussian = np.linspace(-0.5, 0.5, 100)
pdf_gaussian = norm.pdf(x_gaussian)
plt.figure(figsize=(6, 4))
plt.plot(x_gaussian, pdf_gaussian)
plt.title('Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('PDF')
plt.show();
```

## **Exercice 5**

```python
house_sizes = [50, 70, 80, 100, 120]
house_prices = [150000, 200000, 210000, 250000, 280000]
```

```python
# Performing linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

# The slope and intercept of the regression line
print("Slope:", slope)
print("Intercept:", intercept)
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(house_sizes, house_prices)
plt.plot(house_sizes, slope * np.array(house_sizes) + intercept, color='red')
plt.title('Linear Regression')
plt.xlabel('House Size')
plt.ylabel('House Price')
plt.show();
```

```python
price_estimation = slope * 90 + intercept
print("Estimated price for a house of size 90:", price_estimation)
```

Pour 90m², le prix est d'environ 228 972


### Interpretation
Une augmentation de la dimension de la maison entraîne une augmation du prix, donc plus la surface augmente, plus le prix aussi augmente
