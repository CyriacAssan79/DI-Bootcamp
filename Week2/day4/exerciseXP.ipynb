## **Exercice 2**

```python
X = ["Monday", "Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
Y = [72, 74, 76, 80, 82, 78, 75]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8),dpi=80)

plt.plot(X,Y)
plt.xlabel('Day')
plt.ylabel('Temp (°F)')
plt.title('Représentation de la variation de température de Lundi à Dimanche');
```

## **Exercice 3**

```python
sales = [5000, 5500, 6200, 7000, 7500]
month = ["January", "February", "March", "April", "May"]
```

```python
plt.figure(figsize=(10,8),dpi=80)

plt.bar(month,sales)
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('Visualizing Monthly Sales');
```

## **Exercice 4**

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("Student Mental health.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(10,8),dpi=80)
sns.histplot(data=df['What is your CGPA?'],color='red')
plt.title('Visualizing the Distribution of CGPA');
```

## **Exercice 5**

```python
plt.figure(figsize=(10,8),dpi=80)
sns.barplot(data=df, x='Choose your gender', hue='Do you have Anxiety?',palette='Blues');
plt.title('Anxiety Levels Across Different Genders')
```
