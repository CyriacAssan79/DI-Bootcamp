## **Business Report**

### **1. Data Scoping and Preparation**

Nos données ont été chargé depuis kaggle. Ce jeu de donnée comporte 21 colonnes : Order ID, Order Date, Ship Date, Ship Mode, Customer ID, Customer Name, Segment, Country, City, State, Postal Code, Region, Product ID, Category, Sub-Category, Product Name, Sales, Quantity, Discount, Profit.  
De plus ce dataset contient 9994 données.  
  
Etant donnée que ces données ont été récupérer sur leur forme brute, nous allons les prétraiter.

#### **🧹 Handle missing values and duplicates**

Dans notre dataset, nous avons vérifié s'il y avait les valeurs nulles. Après notre vérication, on a constater que toutes les valeurs des colonnes étaient présentes, donc on avait eu aucun prétraitrement à faire ici.

Ensuite nous avons vérifier, s'il y avait des doublons dans les données, c'est-à-dire des valeurs qui se répète sur plusieurs lignes, et il n'y avait pas doublon, donc nous somme passer à la prochaine étape.

#### **🕓 Fix data types**

Dans nos données, afin de faire une bonne analyse avec les dates nous devons les convertirs au format date car dans les données brutes receuillies, Ship Date et Order Date était sous forme de chaine de charactère, pas idéal pour nos analyse se base sur les dates plus tard

#### **🏗️ Feature engineering**

Dans notre Dataset, nous avons ajouté certaines colonnes afin de faire diversifier les informations comme le Profit Margin, Order Year, Order Month et Order Year-Month. Toutes ces variables varibles nous seront utileslors de l'analyse en profondeur de nos données.

### **4. Methodology and Tooling Review**

#### **Analyse comparative des bibliothèques :**  

**=== ANALYSE COMPARATIVE DES BIBLIOTHÈQUES ===**  

POINTS FORTS DE MATPLOTLIB (d'après notre analyse) :
* Contrôle précis des widgets interactifs
* Annotations personnalisées et positionnement du texte
* Mise en page précise des sous-graphiques et dimensionnement des figures
* Intégration avec ipywidgets pour des mises à jour dynamiques

POINTS FORTS DE SEABORN (d'après notre analyse) :

* Visualisations statistiques intégrées (regplot)
* Palettes de couleurs et légendes automatiques
* Style par défaut clair et prêt pour la publication
* Visualisation aisée des données catégorielles

COMPARATIF DE VITESSE :
* Graphique de base Matplotlib : 0,0245 seconde
* Équivalent Seaborn : 0,2764 seconde

Pour une exploration rapide, j'utiliserai Matplotlib car il offre un rendu plus rapide pour les graphiques de base et une intégration fluide avec les widgets interactifs pour une analyse dynamique.
Pour les présentations destinées aux parties prenantes, je privilégierai Seaborn car il offre une esthétique professionnelle, des fonctionnalités statistiques intégrées et des schémas de couleurs professionnels qui améliorent la communication avec la direction.
