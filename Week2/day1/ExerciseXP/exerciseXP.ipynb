# **Exercice 1**


L'analyse des données est l'ensemble des processus permettant de collecter, nettoyer, analyser, interpréter et restituer les données aux équipes métiers afin d'être un support d'aide à la prise de décision. Chaque étape de l'analyse comprend des processus différents :  
  
**La collecte des données** : Il s'agit de recueillir toutes les informations utiles pour mener à bien l'analyse. Ces informations peuvent provenir de différentes sources (sites web, rapports, fichiers, études de marché, etc). De plus ces données extraites peuvent être stockées dans plusieurs fichiers différents formats (csv, log, excel, image, texte etc)  
**Le nettoyage des données**: Lors de la collecte, les données sont recueillies de manière brute et de fois de manière désorganisée. Le processus de nettoyage consiste à, par exemple, supprimer les doublons ou encore extraire les données utiles à l'analyse.  
**L'analyse des données** : Ce processus consiste à vérifier la relation entre les variables explicatives, ou encore identifier les tendances.  
**L'interprétation des résultats** : Ici on tire des conlusions sur le pourquoi d'une action et on resout à ce problème.  
**La restitution** : Dans un projet d'analyse de données, plusieurs acteurs peuvent intervenir, après l'analyse et l'interprétation des données recueillies, les résultats sont transmis aux équipes métiers afin qu'ils puissent prendre des décisions.  
  
Dans l'ère moderne, ce sont des millions de données qui sont traitées chaque jour, l'exploration de ces données devient donc cruciale pour les prises de décision rapide. Cette analyse peut permettre les prises de décision rapide, permet d'avoir un avantage concurrentiel, de comprendre mieux les besoins du client et de gérer les risques.  
  
L'analyse de données peut être utilisée dans plusieurs secteurs, la banque avec la detection de fraute, la sécurité routière avec la détection des zones à risque etc

## **Exercice 3 : Identification des types de données**

### **Rappel**

**Données qualitatives** : Ces données se concentrent sur les descriptions, les opinions, les émotions et les comportements. Elles permettent de comprendre le contexte, les motivations profondes ou les mécanismes derrière un phénomène. On peut avoir des informations textuelles (Non numériques), des questions posées, etc.      
**Données quantitatives**: Ces données se concentrent sur les chiffres, les statistiques et les mesures. Elles permettent de quantifier un phénomène et d'analyser des tendances sur de grands échantillons. On peut avoir des données numériques, et des questions posée ou peut chercher à savoir (Combien, À quelle fréquence, etc.)

## **Exercice 7**

**Une série d'articles de blog sur les expériences de voyage** : On peut uiliser Traitement Automatique du Langage Naturel (NLP) et le text mining pour extraire automatiquement le titre, le pays, la ville, la date, les sentiment, l'activités.  
Les articles de blog sont des textes libres. Le NLP permet d’identifier automatiquement des éléments utiles (lieux, dates, émotions, activités) afin de les stocker dans une base de données exploitable pour l’analyse ou la recommandation.  
**Enregistrements audio des appels au service client** : Utilisation de la reconnaissance vocale (Speech-to-Text) puis appliquer du NLP pour convertir l’audio en texte puis structurer les données en fonction du client, du problème, de la durée, du type de résolution.  
Les données audio ne sont pas directement exploitables par les systèmes analytiques. La transcription transforme la voix en texte, puis les outils NLP permettent d’extraire des informations utiles pour améliorer le service client.  
**Notes manuscrites issues d'une séance de brainstorming** : Utilisation de l’OCR (Reconnaissance Optique de Caractères) associé à l’analyse de texte pour transformer les notes manuscrites en texte numérique structuré, on peut receuillir l'idée principale, la catégorie, les priorités.  
Les notes manuscrites sont difficiles à exploiter automatiquement. L’OCR permet de convertir l’écriture en texte numérique, ensuite organisé en catégories pour faciliter le suivi des idées et la gestion de projet.  
**Un tutoriel vidéo sur la cuisine**: Utilisation de la vision par ordinateur et la transcription audio pour extraire les étapes, les ingrédients, l'action, le temps mis.  
Une vidéo contient plusieurs types de données (image, son, texte). Combiner vision par ordinateur et transcription audio permet d’identifier les ingrédients, les actions et les étapes afin de créer une recette structurée et consultable facilement.
