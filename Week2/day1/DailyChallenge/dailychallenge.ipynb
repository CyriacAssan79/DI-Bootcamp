### **Étape 2**

L'analyse de donnée a été essentiel afin de comprendre les comportements influençant les accidents chez les conducteurs de véhicules de transport collectif informel.  

L’étude adopte une approche qualitative basée sur des entretiens avec conducteurs des transports du secteur informel. Les données recueillies ont été analysé afin d’identifier les comportements de conduite à risque et leurs causes.  

Voici les principaux résultats obtenu :  
•	L’excès de vitesse constitue la principale cause des accidents de la route avec de 9,2% suivie par la conduite dangereuse (7,2%)  
•	Certains comportements à risque sont liés à l’état physique des conducteurs :
>> o	Compétences des conducteurs insuffisant (4,4 %)   
>> o	Consommation de substances psychoactives (4,4 %)


### **Étape 3**

Dans cette étude, l'analyse de données a été essentielle car sans avoir analysé les données, on ne pouvait pas savoir les causes réelles des accidents de circulation. De plus, l'analyse de données a été fondamentale dans cette étude car ils ont découvert les facteurs influençant les accidents. Ces résultats obtenus pourraient constituer un support d'aide à la prise de décision des autorités locales.

### **Étape 4**

En Côte d'Ivoire, il y a plus de 2000 accidents de la circulation selon l'OSER. Dans un premier temps, sans avoir analysé les données, il est difficile de trouver les raisons ou les causes des accidents, ce qui empêche fortement l'intervention des autorités locales. Grâce à la collecte et au traitement des données, on peut facilement vérifier, par exemple les zones fortement exposées, voir si la densité de la population avait un impact sur ces accidents. On peut alors tirer plusieurs conclusions afin de déterminer celles idéales à notre étude et apporter des solutions.
