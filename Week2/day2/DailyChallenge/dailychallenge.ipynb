## **Défi**

```python
import pandas as pd
```

```python
df = pd.read_csv('datascience_salaries.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
```

```python
df['salary'] = mms.fit_transform(df[['salary']])
```

```python
df.head()
```

```python
from sklearn.decomposition import PCA
```

Vu que le PCA fonctionne sur des données numériques, on va scinder nos données

```python
df_num = df.select_dtypes(include= ['int64','float64'])
```

```python
df_num.head()
```

```python
# vérifions s'il y a des valeurs manquantes et traitons les
df_num.isnull().sum()
```

Il n'y a pas de valeurs manquantes donc on peut passer à l'étape suivante:

Application du PCA

```python
pca = PCA(n_components=2)
```

```python

p_composantes = pca.fit_transform(df_num)
```

```python
p_composantes
```

```python
#Créons un nouveau Dataframe pour stocker nos composantes :
p_composante_dataset = pd.DataFrame(p_composantes, columns=['PC1','PC2'])
```

```python
p_composante_dataset.head()
```

```python
#Regroupons l'ensemble de données par la colonne « niveau_d'expérience » et calculez le salaire moyen et médian pour chaque niveau d'expérience (par exemple, junior, intermédiaire, senior).
```

```python
df.groupby('experience_level')['salary'].mean()
```

On remarque que les seniors et les cadres (executive) ont un salaire moyen supérieur à celle des débutants et des intermédiaires.

```python
df.groupby('experience_level')['salary'].median()
```
