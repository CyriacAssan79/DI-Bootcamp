# Exercice 4 : Traitement des valeurs aberrantes (Outliers)

Dans cette section, nous identifions les valeurs extrêmes pour les colonnes `Fare` (Tarif) et `Age` afin d'éviter qu'elles ne biaisent nos futurs modèles statistiques.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 8), dpi=80)
sns.boxplot(y='Age', data=df_train)
plt.xticks(rotation=90);
```

Dans le dataset, on abserve que parmis les ages des passager, certaines ages sont éloignées du regroupement, ces personnes pourraient fausser nos analyses

```python
plt.figure(figsize=(10, 8), dpi=80)
sns.boxplot(y = df_train['Fare'])
plt.xticks(rotation=90);
```

Pour le tarif, vu que les prix des billets ne sont pas les mêmes, d'autre peux payer des billets hyper cher et d'autre moins chère, dans l'analyse de données un écart trop grand entre les valeurs pourrait entrainer des anlyses biaisés

```python
# Identify and handle outliers in 'Fare' column
Q1 = df_train['Fare'].quantile(0.25)
```

```python
Q1
```

On remarque 25% des passagers payent en moyenne 7,87

```python
Q3 = df_train['Fare'].quantile(0.75)
```

```python
Q3
```

75% des passagers ont payé en moyenne 31$

```python
IQR = Q3-Q1
```

```python
IQR
```

Ensuite; on doit définir les bornes supérieures et inférieures pour pouvoir supprimer les valeurs aberrantes lorsqu'elles atteignent le seuil de supériorité ou d'infériorité.

```python
# Bornes classiques
borne_inf = Q1 - 1.5 * IQR
borne_sup = Q3 + 1.5 * IQR
```

```python
# Récupération des outliers
outliers = df_train[
    (df_train['Fare'] < borne_inf) |
    (df_train['Fare'] > borne_sup)
]
```

```python
print("Nombre d'outliers :", outliers.shape[0])
```

Il y a 114 outliers pour la colonne 'Fare'.

```python
df_sans_outliers = df_train[
    (df_train['Fare'] >= borne_inf) &
    (df_train['Fare'] <= borne_sup)
]
```

```python
plt.figure(figsize=(10, 8), dpi=80)
sns.boxplot(y='Fare', data=df_sans_outliers)
plt.xticks(rotation=90);
```

Tous les outliers ont été supprimer pour le prix

```python
# Calcul des bornes spécifiques pour l'Age
Q1_age = df_train['Age'].quantile(0.25)
Q3_age = df_train['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age

borne_inf_age = Q1_age - 1.5 * IQR_age
borne_sup_age = Q3_age + 1.5 * IQR_age

# Récupération des outliers pour Age
outliers_age = df_train[
    (df_train['Age'] < borne_inf_age) |
    (df_train['Age'] > borne_sup_age)
]
```

```python
print("Nombre d'outliers :", outliers_age.shape[0])
```

```python
# Suppression des outliers pour l'Age dans df_train
df_train = df_train[
    (df_train['Age'] >= borne_inf_age) &
    (df_train['Age'] <= borne_sup_age)
]

print("Nouvelle taille du dataset :", df_train.shape[0])
```

```python
plt.figure(figsize=(10, 8), dpi=80)
sns.boxplot(y='Age', data=df_train)
plt.xticks(rotation=90);
```

Pour la variable Age, tous les outliers n'ont pas été supprimé mais ils ont été considérablement réduit

```python
# Affichage du bilan final
print(f"Nombre de lignes initial : 891")
print(f"Nombre de lignes après suppression des outliers : {df_train.shape[0]}")
print(f"Total de lignes supprimées : {891 - df_train.shape[0]}")
```

### À noter

Une caractéristique est normalement distribué si :  
*  la courbe KDE a une forme de cloche
les données sont concentrées autour de la moyenne
*  les valeurs diminuent progressivement de chaque côté
*  la distribution est relativement symétrique

Une caractéristique est normalement distribué si :  
*  une queue est plus longue d’un côté
*  les valeurs ne sont pas réparties de manière équilibrée
*  il existe beaucoup de petites valeurs et quelques très grandes valeurs



```python
sns.histplot(df_train['Fare'], kde=True)

plt.show()
```

La varibale Fare est asymétrique car une queue est plus longue d'un côte que l'autre et les valeurs ne sont pas répartis de façon équilibré

La fonction **pd.cut** de Pandas permet de convertir une variable continue (numérique) en variable catégorielle en segmentant et triant les valeurs dans des intervalles spécifiques

```python
limites = [0, 12, 18, 60, 100]
noms_categories = ['enfant', 'adolescent', 'adulte', 'senior']

df_train['Age_life'] = pd.cut(df_train['Age'], bins=limites, labels=noms_categories)
```
