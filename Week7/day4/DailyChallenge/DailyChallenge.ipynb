## **1. Comprendre l'évaluation du LLM**

#### **Pourquoi l'évaluation des LLM est-elle plus complexe ?**

Les LLM sont non déterministes : une même question peut avoir plusieurs réponses correctes. Leur évaluation porte aussi sur la pertinence, la cohérence, la sécurité et la qualité du langage, contrairement aux logiciels classiques où les résultats attendus sont fixes.

#### **Pourquoi évaluer la sécurité d'un LLM ?**
* Éviter les hallucinations et la désinformation.
* Limiter les contenus dangereux ou offensants.
* Protéger les données sensibles.
* Réduire les biais.
* Garantir une utilisation fiable et responsable.
#### **Rôle des tests contradictoires**

Ils soumettent le modèle à des entrées difficiles ou malveillantes pour :

* détecter ses faiblesses ;
* améliorer sa robustesse ;
* renforcer ses mécanismes de sécurité.
#### **Limites des métriques automatiques vs évaluation humaine**
* Métriques automatiques (BLEU, ROUGE, perplexité) : rapides et objectives, mais ne mesurent pas toujours la pertinence, le raisonnement ou la véracité.
* Évaluation humaine : plus complète (qualité, contexte, utilité), mais plus lente, coûteuse et parfois subjective.

En somme, la meilleure approche combine métriques automatiques, évaluation humaine et tests contradictoires.

## **2. Application des indicateurs BLEU et ROUGE**

```python
import nltk
from nltk.translate.bleu_score import sentence_bleu
from nltk.translate.bleu_score import SmoothingFunction

# BLEU Score Calculation
reference_bleu = ["Malgré le recours croissant à l’intelligence artificielle dans divers secteurs, la supervision humaine demeure essentielle pour garantir une mise en œuvre éthique et efficace.".split()]
generated_bleu = "Bien que l'IA soit de plus en plus utilisée dans l'industrie, la supervision humaine reste nécessaire pour une application éthique et efficace.".split()

# Using SmoothingFunction().method1 for a better score when short sentences or few overlaps
chencherry = SmoothingFunction()
bleu_score = sentence_bleu(reference_bleu, generated_bleu, smoothing_function=chencherry.method1)
print(f"BLEU Score (with smoothing): {bleu_score:.4f}")
```

```python
from rouge_score import rouge_scorer

# ROUGE Score Calculation
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

reference_rouge = "Face à l’évolution rapide du climat, les initiatives mondiales doivent se concentrer sur la réduction des émissions de carbone et le développement de sources d’énergie durables afin d’atténuer l’impact environnemental."
generated_rouge = "Pour lutter contre le changement climatique, les efforts mondiaux devraient viser à réduire les émissions de carbone et à favoriser le développement des énergies renouvelables."

scores = scorer.score(reference_rouge, generated_rouge)

print("ROUGE Scores:")
for key, value in scores.items():
    print(f"{key}: P={value.precision:.4f}, R={value.recall:.4f}, F={value.fmeasure:.4f}")
```

### Limites des systèmes BLEU et ROUGE pour l'évaluation de textes créatifs ou contextuels

BLEU et ROUGE sont limités car ils se basent sur la correspondance lexicale (n-grammes) avec une référence fixe. Ils ne mesurent pas :

*   **Le sens sémantique** : Ignorent les synonymes et la compréhension réelle.
*   **La pertinence contextuelle** : Ne peuvent pas évaluer l'adéquation au contexte global.
*   **La fluidité et la grammaire** : Pas conçues pour la qualité grammaticale ou stylistique.
*   **La véracité** : Ne détectent pas les faits inventés ou les hallucinations.
*   **La créativité ou l'originalité** : Récompensent la ressemblance plutôt que la nouveauté.

Ils pénalisent les paraphrases valides et l'originalité, ce qui les rend inadaptés pour évaluer des textes créatifs ou nuancés.

### Améliorations ou méthodes alternatives pour évaluer la génération de texte

Pour une meilleure évaluation, on utilise :

1.  **Évaluation Humaine** : La méthode la plus fiable pour la nuance, le sens et la créativité, malgré son coût et sa subjectivité. Elle inclut le classement, la notation absolue ou les tests basés sur les tâches.
2.  **Métriques Basées sur les Embeddings (ex. BERTScore)** : Mesurent la similarité sémantique, plus robustes aux paraphrases que BLEU/ROUGE.
3.  **LLM-as-a-Judge** : Un LLM évalue le texte généré (notation, classement), offrant rapidité et évolutivité mais avec des risques de biais.
4.  **Métriques de Cohérence** : Évaluent la fluidité et la logique (ex. perplexité contextuelle).
5.  **Métriques de Sécurité et d'Éthique** : Détectent le contenu toxique, les biais ou les vulnérabilités.

La meilleure approche combine ces méthodes, notamment les métriques automatiques et l'évaluation humaine.

## **3. Analyse de perplexité**

**Comparaison de la perplexité**

Le modèle A présente la plus faible perplexité, car il attribue une probabilité plus élevée (0,8) au mot correct que le modèle B (0,4).

Pourquoi ? Une probabilité plus élevée signifie que le modèle est plus confiant et prédit mieux le mot attendu, ce qui réduit la perplexité.

**Perplexité de 100 : implications**

Une perplexité de 100 indique que le modèle prédit relativement mal les mots suivants. Ses performances sont donc limitées, avec un risque plus élevé d'erreurs et de réponses moins cohérentes.

Pour l'améliorer :

* utiliser un jeu de données plus riche et de meilleure qualité ;
entraîner davantage le modèle (fine-tuning) ;
* augmenter la taille ou la capacité du modèle ;
* améliorer le prétraitement et le nettoyage des données.

## **5. Exercice de test contradictoire**

**Erreur potentielle**

L'étudiant pourrait comprendre autrement la question et donner une réponse incorrecte ou complexe (par exemple le siège d'une organisation), alors que la réponse attendue est simplement Paris.  

**Méthode pour améliorer la robustesse**  

* Tester le modèle avec des questions simples, ambiguës et reformulées.
* Vérifier automatiquement les réponses à l'aide de sources fiables.
* Utiliser des jeux de tests contradictoires variés pour détecter les erreurs.  

**Trois questions pièges**  

* Combien y a-t-il de jours dans une année bissextile ?  
-> Réponse attendue : 366.
* Le Soleil tourne-t-il autour de la Terre ?  
-> Réponse attendue : Non, la Terre tourne autour du Soleil.  
* Quelle est la capitale de la Côte d'Ivoire (sans précisé économique ou politique)?

## **6. Analyse comparative des méthodes d'évaluation**

Tâche choisie : Réponse aux questions (Question Answering)  
|Métrique	|Avantages |	Limites|
|:---|:---|:---|
|BLEU|	Mesure le chevauchement des mots avec la réponse de référence|	Pénalise les reformulations et ne mesure pas le sens|
|BERTScore |Compare le sens sémantique des réponses, même avec des formulations différentes|Plus coûteux à calculer et dépend d'un modèle pré-entraîné|
|Évaluation humaine	|Évalue la pertinence, l'exactitude, la cohérence et la clarté de la réponse|Lente, coûteuse et parfois subjective|

**Métrique la plus appropriée**

Pour une tâche de réponse aux questions, BERTScore est la métrique la plus adaptée, car elle évalue la similarité sémantique plutôt que le simple chevauchement de mots. Deux réponses peuvent être formulées différemment tout en étant correctes. Il est toutefois recommandé de la compléter par une évaluation humaine afin de vérifier l'exactitude factuelle et la qualité globale des réponses.
