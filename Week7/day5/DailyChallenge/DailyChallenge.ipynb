### **Générer la réponse du modèle**

```python
import google.generativeai as genai
from google.colab import userdata

# Configure Gemini API
# Make sure you have your GOOGLE_API_KEY stored in Colab secrets
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

# Initialize the Generative Model
gemini_model = genai.GenerativeModel('gemini-2.5-flash')

input_text = """Les employés doivent s'assurer que tout accès distant aux systèmes internes est établi via le VPN sécurisé approuvé. En aucun cas, des connexions non sécurisées ou des appareils personnels dépourvus de protection des terminaux ne doivent être utilisés pour accéder à des données confidentielles ou à des communications sensibles."""

instruction = """Agis comme un spécialiste de la communication interne d'une entreprise technologique. Réécris le texte ci-dessous pour une newsletter de micro-apprentissage destinée aux employés. Utilise un ton amical, simple et facile à comprendre. Organise les informations sous forme de puces. Paraphrase entièrement le contenu sans reprendre de phrases ou d'expressions du texte original. La réponse finale doit contenir moins de 75 mots et ne doit inclure aucune information qui ne figure pas dans le texte source."""

full_prompt = f"""Texte d'entrée :\n-> {input_text}\n\nInstruction :\n-> {instruction}\n\nSortie :"""

response = gemini_model.generate_content(full_prompt)

print(response.text)
```

Ensuite, nous discuterons du type de communication où une citation serait plus appropriée et des risques potentiels des citations.

```python
instruction_quote = """Agis comme un expert en analyse de politiques de sécurité. À partir du texte fourni, identifie la phrase unique qui représente le mieux la politique de sécurité essentielle et extrais-la mot pour mot. La réponse doit être uniquement la citation, sans aucune introduction, explication ou ponctuation additionnelle. La citation doit être exacte."""

full_prompt_quote = f"""Texte d'entrée :\n-> {input_text}\n\nInstruction :\n-> {instruction_quote}\n\nSortie:"""

response_quote = gemini_model.generate_content(full_prompt_quote)

print(response_quote.text)
```

### Analyse de la citation

**Dans quel type de communication interne la citation serait-elle plus appropriée que la paraphrase ?**

Une citation directe serait plus appropriée dans des communications internes où l'exactitude juridique ou politique est primordiale, par exemple :
*   **Documents officiels de politique** : Pour s'assurer que la formulation exacte est connue et respectée.
*   **Formations de conformité** : Lorsque les employés doivent comprendre la formulation précise pour des questions de responsabilité.
*   **Communications légales ou RH** : Pour éviter toute ambiguïté ou interprétation erronée.
*   **Tableaux d'affichage critiques ou rappels de sécurité** : Où l'impact direct et la clarté concise d'une règle spécifique sont nécessaires.

**Dans quelles circonstances les devis peuvent-ils présenter un risque ?**

Les citations peuvent présenter des risques dans les circonstances suivantes :
*   **Manque de contexte** : Une citation sortie de son contexte peut être mal interprétée ou sembler plus sévère qu'elle ne l'est.
*   **Jargon technique/juridique** : Si la citation contient un langage complexe, elle peut être difficile à comprendre pour un public non spécialisé.
*   **Longueur excessive** : Une longue citation peut décourager la lecture et la compréhension, en particulier dans les communications rapides.
*   **Ton inadapté** : Le ton de la citation originale peut ne pas correspondre au ton souhaité pour la communication, rendant le message moins engageant ou plus intimidant.
*   **Perception négative** : Les citations peuvent être perçues comme rigides ou autoritaires, ce qui peut nuire à l'engagement des employés par rapport à une paraphrase plus conviviale.
