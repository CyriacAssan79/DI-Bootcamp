### **Chargement du jeu de données**

```python
%pip install -Uq datasets --quiet
```

```python
dataset_name = "databricks/databricks-dolly-15k"
page_content_column = "context"
```

### **Création d'une instance HuggingFaceDatasetLoader et chargement des données**

```python
from langchain_community.document_loaders import HuggingFaceDatasetLoader
```

```python
loader = HuggingFaceDatasetLoader(dataset_name, page_content_column)
data = loader.load()
print(data[:2]) # Optional: Print the first 2 entries to verify loading
```

### **Fractionnons les documents**

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=150)
```

```python
docs = text_splitter.split_documents(data)
print(docs[0])
```

### **Intégration du texte**

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
modelPath = "sentence-transformers/all-MiniLM-l6-v2"
model_kwargs = {'device':'cpu'}
encode_kwargs = {'normalize_embeddings': False}
```

```python
#Initialisation de HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(
  model_name=modelPath,
  model_kwargs=model_kwargs,
  encode_kwargs=encode_kwargs
)
```

```python
#Création de l'intégration de test :
text = "This is a test document."
query_result = embeddings.embed_query(text) #Vectorisation du mot
print(query_result[:3])
```

### **Création d'un entrepôt de vecteurs**

```python
from langchain_community.vectorstores import FAISS
```

```python
db = FAISS.from_documents(docs, embeddings)
```

## **Préparation du modèle LLM**

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, pipeline
from langchain_community.llms import HuggingFacePipeline
```

```python
#Chargement du tokenizer et du modèle de question-réponse
tokenizer = AutoTokenizer.from_pretrained("Intel/dynamic_tinybert")
model = AutoModelForQuestionAnswering.from_pretrained("Intel/dynamic_tinybert")
```

```python
#Création d'un processus de questions-réponses :

model_name = "Intel/dynamic_tinybert"
tokenizer = AutoTokenizer.from_pretrained(model_name, padding=True, truncation=True, max_length=512)
Youtubeer = pipeline(
  "question-answering",
  model=model_name,
  tokenizer=tokenizer,
  return_tensors='pt'
)
```

```python
#Création d'un wrapper de pipeline Langchain :

llm = HuggingFacePipeline(
  pipeline=Youtubeer,
  model_kwargs={"temperature": 0.7, "max_length": 512},
)
```

## **Création de la chaîne de questions-réponses pour la recherche**

```python
from langchain_community.chains import RetrievalQA
```

```python
retriever = db.as_retriever(search_kwargs={"k": 3}) # Optional: You can adjust k for number of documents retrieved
```

```python
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="refine", retriever=retriever, return_source_documents=False)
```

## **Testons notre système RAG**

```python
question = "What is cheesemaking?"
```

```python
result = qa.run({"query": question})
print(result) # Or print(result["result"]) if the output is a dictionary
```
